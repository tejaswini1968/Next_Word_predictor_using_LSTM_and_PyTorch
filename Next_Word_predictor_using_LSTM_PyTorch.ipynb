```python
#conveting Text to numerical indices
def text_to_indices(sentence, vocab):

  numerical_sentence = []

  for token in sentence:
    if token in vocab:
      numerical_sentence.append(vocab[token])
    else:
      numerical_sentence.append(vocab['<UNK>'])

  return numerical_sentence
```

```python
#defination of custom dataset
class CustomDataset(Dataset):

    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, index):
        return self.X[index], self.y[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
#defination of data loader
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
len(dataloader)
```

```python
# shape of training batches
for batch_x, batch_y in dataloader:
    print(batch_x.shape)
    print(batch_y.shape)
    break
```

```python
#defination of LSTM model architecture
class LSTMModel(nn.Module):

    def __init__(self, vocab_size):
        super().__init__()

        #converts word indices to dense vectors
        self.embedding = nn.Embedding(vocab_size, 100,padding_idx=0)

        #LSTM layer
        self.lstm = nn.LSTM(input_size=100, hidden_size=150, num_layers=2,dropout=0.3,batch_first=True)

        #final classification
        self.fc = nn.Linear(150, vocab_size)

    def forward(self, X):
        #shape(batch_size,seq_layer)
        embedded = self.embedding(X)

        #output shape(batch_size,seq_layer,100)
        #_,(hidden,_)=self.lstm(embedded)
        intermediate_hidden_states, (final_hidden_state, final_cell_state) = self.lstm(embedded)

        output = self.fc(final_hidden_state[-1])

        #shape:(batch_size,vocab_size)
        return output
```

```python
model = LSTMModel(len(vocab))
```

```python
#model summary
summary(model, input_size=(32, X.shape[1]), dtypes=[torch.long])
```

```python
model.to(device)
```

```python
#defination of hyper parameter
epochs = 50
learning_rate = 0.001

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
#training loop
for epoch in range(epochs):
    model.train()
    total_loss = 0

    for batch_x, batch_y in dataloader:
        #move data to GPU/CPU
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)

        #clear old gradients
        optimizer.zero_grad()
        #forward pass
        output = model(batch_x)

        #compute loss
        loss = criterion(output, batch_y)

        #backpropogation
        loss.backward()

        #update weights
        optimizer.step()

        total_loss += loss.item()

        #Accumulate loss
    avg_loss = total_loss / len(dataloader)

    print(f"Epoch {epoch + 1} | Loss {avg_loss:.4f}")
```

```python
#Swapping the vocab dictoinary from "word to index" -> "index to word form"
index_to_word = {index: word for word,index in vocab.items()}
index_to_word
```

```python
#prediction function
def predict(model, vocab, text):
    model.to(device)
    model.eval()

    tokenized_text = word_tokenize(text.lower())
    numerical_text = text_to_indices(tokenized_text, vocab)

    #sequence length used during training
    max_length = X.shape[1]
    #keep only lastmax_length  words ,if input sentennce is shorter than seq length
    #adds left if input sentennce is shorter than seq length
    numerical_text = numerical_text[-max_length:]
    padded_text = [0] * (max_length - len(numerical_text)) + numerical_text
    #coverts to tensor and adds batch dimension
    padded_text_tensor = torch.tensor(padded_text, dtype=torch.long).unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(padded_text_tensor)
        _, index = torch.max(output, dim=1)

    predicted_word = index_to_word[index.item()]
    return text + " " + predicted_word
```

```python
#auto regressive text generation
def prediction( num_tokens:int=None, input_text:str =None) -> None:

    for i in range(num_tokens):
        output_text = predict(model, vocab, input_text)
        print(output_text)
        input_text = output_text
        time.sleep(0.5)
```

```python
# generating ouput in single step
def prediction_full_text(num_tokens:int = None, input_text:str = None):

    generated_text = input_text

    for _ in range(num_tokens):
        generated_text = predict(model,vocab,generated_text)

    return generated_text
```

```python
#accuracy on training data
model.eval()
total = 0
correct = 0

with torch.no_grad():
  for batch_x, batch_y in dataloader:
    batch_x, batch_y = batch_x.to(device), batch_y.to(device)

    output = model(batch_x)
    _, predicted = torch.max(output, dim=1)

    correct += (predicted == batch_y).sum().item()
    total += batch_y.size(0)



accuracy = ( correct / total ) * 100
print(f"Model Accuracy:{accuracy:.4f}")

```
